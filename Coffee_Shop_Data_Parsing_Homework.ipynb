```python
BASE_URL = "https://colleenvanlent.github.io/101_coffee_shops/api"

import hashlib
import random
import re
import requests

def _dataset_number_from_uniqname(u):
    u = (u or "").strip().lower()
    if u == "":
        raise ValueError("You must set your uniqname before running this cell.")
    digest = hashlib.sha256(u.encode("utf-8")).hexdigest()
    return (int(digest, 16) % 8) + 1   # 1–8

def _extract_shop_list(payload):
    for k, v in payload.items():
        if k in ("id", "last_updated"):
            continue
        if isinstance(v, list) and (len(v) == 0 or isinstance(v[0], dict)):
            return v, k
    raise KeyError(f"Could not find the shop list. Keys: {sorted(payload.keys())}")

def _remix_date_format(date_str):
    """
    Take something like '2026-02-07' or '2026-2-7' (or similar),
    extract year/month/day, then randomly reformat month/day with
    1- or 2-digit padding.
    """
    if not isinstance(date_str, str):
        return date_str  # leave it alone if missing/None/etc.

    m = re.match(r"^\s*(\d{4})-(\d{1,2})-(\d{1,2})\s*$", date_str)
    if not m:
        return date_str  # if format is unexpected, don't break anything

    year = m.group(1)
    month = int(m.group(2))
    day = int(m.group(3))

    pad_month = random.choice([True, False])
    pad_day = random.choice([True, False])

    month_str = f"{month:02}" if pad_month else str(month)
    day_str = f"{day:02}" if pad_day else str(day)

    return f"{month_str}-{day_str}-{year}"

dataset_num = _dataset_number_from_uniqname(uniqname)
dataset_file = f"aa_coffee_exam_{dataset_num}.json"
dataset_url = f"{BASE_URL}/{dataset_file}"

response = requests.get(dataset_url)
response.raise_for_status()

payload = response.json()

# Randomize the formatting AFTER loading (prevents hardcoding assumptions)
payload["last_updated"] = _remix_date_format(payload.get("last_updated", "Unknown"))

coffee_data = payload

print(f"Loaded dataset")

## Feel free to print the data - it is called coffee_data
```
